```python
import matplotlib.pyplot as plt

region_sales = df.groupby("Region")["Sales"].sum()

region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

*Insights*:

1. West region has highest sales.
2. Technology category generates more revenue.
3. Some regions have lower performance compared to others.
